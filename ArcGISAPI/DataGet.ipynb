# Grab land cover for a given parcel
https://developers.arcgis.com/python/guide/raster-analysis-advanced-concepts/

```python
#Import the GIS and display modules
from arcgis import GIS
from IPython.display import display
```

```python
#Create the GIS object, authenticating with your ArcGIS Pro account
gis = GIS('pro')
```

```python
#Search for land cover
lc_results = gis.content.search("NLCD 2011")
len(lc_results)
```

```python
#Get the first result
lc = lc_results[0]
```

```python
#Get the URL
lc.url
```

```python
lc_lyr = lc.layers[0]
lc_lyr
```

```python
lc_sr = lc.spatialReference
```

```python
m = gis.map("Durham, NC")
m
```

```python
from arcgis.geocoding import geocode
area = geocode('Durham, NC',out_sr = lc.spatialReference)[0]
area['extent']
```

```python
from IPython.display import Image
```

```python
img = lc_lyr.export_image(bbox=area['extent'],area['extent'])
```
