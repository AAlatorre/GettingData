### The *Content Manager* ([link](https://esri.github.io/arcgis-python-api/apidoc/html/arcgis.gis.toc.html#contentmanager))
The Content Manager is a helper class for managing content in ArcGIS Online or ArcGIS Enterprise. This class is not created by users directly. An instance of this class, called ‘content’, is available as a property of the GIS object. Users call methods on this ‘content’ object to manipulate (create, get, search, etc) items.

Among the operations included in the `content` sub-module is [`search`](https://esri.github.io/arcgis-python-api/apidoc/html/arcgis.gis.toc.html#arcgis.gis.ContentManager.search), which enables us to search for resources hosted on AGOL. We provide a [query](https://developers.arcgis.com/rest/users-groups-and-items/search-reference.htm), in dictionary format, and the search operation returns a list of resources meeting that criteria. Refer to the link for a complete list of search terms/keys.
